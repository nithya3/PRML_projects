```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
os.listdir('Dataset')
```

```python
ratings = pd.read_csv('Dataset/ratings.csv')
remarks = pd.read_csv('Dataset/remarks.csv')
```

```python
remarks.rename({'remarkDate':'Date'}, axis=1, inplace=True)
```

```python
remarks_ratings = ratings.merge(remarks, how='outer')
```

```python
remarks_ratings.replace(np.nan, '', inplace = True)
```

```python
remarks_ratings['rating'].replace('', np.nan, inplace = True)
```
