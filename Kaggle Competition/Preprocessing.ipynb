```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from datetime import datetime
```

```python
os.listdir('Dataset')
```

```python
ratings = pd.read_csv('Dataset/ratings.csv')
remarks = pd.read_csv('Dataset/remarks.csv')
remarks_supp_opp = pd.read_csv('Dataset/remarks_supp_opp.csv')
train = pd.read_csv('Dataset/train.csv')
```

```python
remarks.rename({'remarkDate':'Date'}, axis=1, inplace=True)
```

```python
remarks_ratings = ratings.merge(remarks, how='outer')
```

```python
emp_comp = remarks_supp_opp.merge(remarks_ratings, how='outer')
emp_comp.drop(['oppose'], axis = 1, inplace=True)
```

```python
emp_comp.replace(np.nan, '', inplace = True)
emp_comp['rating'].replace('', np.nan, inplace = True)
```

```python
emp_comp
```

```python
emp_comp['Date'] = emp_comp['Date'].apply(lambda d: datetime.strptime(d, '%d-%m-%Y').date() if d != '' else '')
```

##### Overall

```python
dates = sorted(np.unique(emp_comp['Date'].values[np.argwhere(emp_comp['Date'].values != '')]).tolist())
```

```python
day_gaps = []
for i in range(len(dates)-1):
    day_gaps.append((dates[i+1]-dates[i]).days)
day_gaps
```

#### Company-wise

```python
comps = np.unique(emp_comp['comp'].values)
```

```python
for c in comps:
    dates = sorted(np.unique(emp_comp['Date'].values[np.argwhere(emp_comp['Date'].values != '')]).tolist())
    day_gaps = []
    for i in range(len(dates)-1):
        day_gaps.append((dates[i+1]-dates[i]).days)
    print('\n\nCompany: {0}, Avg freq of reviews: {1}'.format(c, sum(day_gaps)/len(day_gaps)))
    print(day_gaps)
```

##### There is no diference in the company-wise analysis
